### Importing all the necessary libraries.

```python
import requests
import IPython.display as disp
from pprint import pprint
```

```python
title = input('Enter Movie Title: ')
url = f"http://www.omdbapi.com/?apikey=16c3139c&t={title}"
resp=requests.get(url)

if resp.status_code == 200:
    print(resp.json())
    
else:
    print('Invallid URL')
```

```python
title = input('Enter Movie Title: ')
url = f"http://www.omdbapi.com/?apikey=16c3139c&t={title}"
resp=requests.get(url)

if resp.status_code == 200:
    data = resp.json()
    show = ['Title', 'Actors', 'Director','Language', 'Genre', 'totalSeasons', 'Ratings', 'Plot']
    img = data["Poster"]
    display(disp.Image(requests.get(img).content))
    
    
    for i in show:
        print(f'{i} ---> {data[i]}')
    
else:
    print('Invallid URL')
```
